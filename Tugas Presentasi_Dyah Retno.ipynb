**Penjelasan Tentang Stunting dan Wasting**

Stunting dan wasting merupakan bentuk kekurangan gizi pada anak, tetapi keduanya mewakili aspek pertumbuhan dan perkembangan yang berbeda.

- Stunting merujuk pada anak yang terlalu pendek untuk usianya akibat kekurangan gizi kronis.

- Wasting merujuk pada anak yang terlalu kurus untuk tinggi badannya, sering kali akibat kekurangan gizi akut atau penyakit.

catatan :

**Kolom Stunting**
- Normal = postur tubuh normal
- Stunted = pertumbuhan terhambat
- Severely stunted = pertumbuhan sangat terhambat
- Tall = postur tubuh tinggi

**Kolom Wasting**
- Normal weight = berat normal
- underweight = kurus
- severely underweight = sangat kurus
- risk of overweight = resiko kelebihan BB
